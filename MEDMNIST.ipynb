```python
import flwr as fl
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split, Subset
import numpy as np
import matplotlib.pyplot as plt
import medmnist
from medmnist import INFO
import torchvision.transforms as transforms

# Dataset setup
DATASET_NAME = "pathmnist"
info = INFO[DATASET_NAME]
DataClass = getattr(medmnist, info["python_class"])

# Device
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Transform
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[.5], std=[.5])
])

# Load MedMNIST data
train_dataset = DataClass(split='train', transform=transform, download=True)
test_dataset = DataClass(split='test', transform=transform, download=True)

# CNN Model
class SimpleCNN(nn.Module):
    def __init__(self, in_channels=1, num_classes=10):
        super(SimpleCNN, self).__init__()
        self.network = nn.Sequential(
            nn.Conv2d(in_channels, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Flatten(),
            nn.Linear(64 * 7 * 7, 128),
            nn.ReLU(),
            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        return self.network(x)

# Client data loader
def get_client_dataloader(client_id, num_clients=5, batch_size=32):
    total_size = len(train_dataset)
    client_size = total_size // num_clients
    indices = list(range(client_id * client_size, (client_id + 1) * client_size))
    subset = Subset(train_dataset, indices)
    train_len = int(0.8 * len(subset))
    val_len = len(subset) - train_len
    train_data, val_data = random_split(subset, [train_len, val_len])
    return (
        DataLoader(train_data, batch_size=batch_size, shuffle=True),
        DataLoader(val_data, batch_size=batch_size, shuffle=False),
    )

testloader = DataLoader(test_dataset, batch_size=32)

# Flower Client
class FlowerClient(fl.client.NumPyClient):
    def __init__(self, cid):
        self.cid = cid
        self.model = SimpleCNN(in_channels=info["n_channels"], num_classes=len(info["label"])).to(DEVICE)
        self.trainloader, self.valloader = get_client_dataloader(cid)

    def get_parameters(self, config):
        return [val.cpu().numpy() for val in self.model.state_dict().values()]

    def set_parameters(self, parameters):
        state_dict = self.model.state_dict()
        for k, v in zip(state_dict.keys(), parameters):
            state_dict[k] = torch.tensor(v)
        self.model.load_state_dict(state_dict)

    def fit(self, parameters, config):
        self.set_parameters(parameters)
        self.model.train()
        optimizer = optim.SGD(self.model.parameters(), lr=0.01)
        criterion = nn.CrossEntropyLoss()
        for x, y in self.trainloader:
            x, y = x.to(DEVICE), y.squeeze().long().to(DEVICE)
            optimizer.zero_grad()
            out = self.model(x)
            loss = criterion(out, y)
            loss.backward()
            optimizer.step()
        return self.get_parameters({}), len(self.trainloader.dataset), {}

    def evaluate(self, parameters, config):
        self.set_parameters(parameters)
        self.model.eval()
        criterion = nn.CrossEntropyLoss()
        correct, loss, total = 0, 0, 0
        with torch.no_grad():
            for x, y in self.valloader:
                x, y = x.to(DEVICE), y.squeeze().long().to(DEVICE)
                out = self.model(x)
                loss += criterion(out, y).item()
                pred = torch.argmax(out, dim=1)
                correct += (pred == y).sum().item()
                total += y.size(0)
        return loss / len(self.valloader), len(self.valloader.dataset), {"val_accuracy": correct / total}

# Client factory
def client_fn(cid: str):
    return FlowerClient(int(cid)).to_client()

# Server-side eval
test_losses, test_accuracies = [], []

def evaluate_fn(server_round, parameters, config):
    model = SimpleCNN(in_channels=info["n_channels"], num_classes=len(info["label"])).to(DEVICE)
    state_dict = model.state_dict()
    for k, v in zip(state_dict.keys(), parameters):
        state_dict[k] = torch.tensor(v)
    model.load_state_dict(state_dict)
    model.eval()

    criterion = nn.CrossEntropyLoss()
    correct, total, loss = 0, 0, 0

    with torch.no_grad():
        for x, y in testloader:
            x, y = x.to(DEVICE), y.squeeze().long().to(DEVICE)
            out = model(x)
            loss += criterion(out, y).item()
            pred = torch.argmax(out, dim=1)
            correct += (pred == y).sum().item()
            total += y.size(0)

    accuracy = correct / total
    avg_loss = loss / len(testloader)
    test_losses.append(avg_loss)
    test_accuracies.append(accuracy)

    print(f"Round {server_round}: Test Loss = {avg_loss:.4f}, Accuracy = {accuracy:.4f}")
    return avg_loss, {"test_accuracy": accuracy}

# Run Federated Simulation
if __name__ == "__main__":
    model = SimpleCNN(in_channels=info["n_channels"], num_classes=len(info["label"]))
    initial_weights = [val.cpu().numpy() for val in model.state_dict().values()]
    initial_parameters = fl.common.ndarrays_to_parameters(initial_weights)

    fl.simulation.start_simulation(
        client_fn=client_fn,
        num_clients=5,
        config=fl.server.ServerConfig(num_rounds=100),
        strategy=fl.server.strategy.FedAvg(
            fraction_fit=1.0,
            fraction_evaluate=1.0,
            min_fit_clients=5,
            min_evaluate_clients=5,
            min_available_clients=5,
            evaluate_fn=evaluate_fn,
            initial_parameters=initial_parameters,
        ),
    )

    # Save metrics
    np.save("medmnist_fedavg_test_accuracies.npy", np.array(test_accuracies))
    np.save("medmnist_fedavg_test_losses.npy", np.array(test_losses))

    # Plot results
    rounds = list(range(1, len(test_losses) + 1))
    fig, ax1 = plt.subplots(figsize=(10, 5))

    ax1.set_xlabel("Round")
    ax1.set_ylabel("Test Loss")
    line1, = ax1.plot(rounds, test_losses, marker='o', label="Test Loss")
    ax1.tick_params(axis='y')

    ax2 = ax1.twinx()
    ax2.set_ylabel("Test Accuracy")
    line2, = ax2.plot(rounds, test_accuracies, marker='o', label="Test Accuracy")
    ax2.tick_params(axis='y')

    lines = [line1, line2]
    labels = [line.get_label() for line in lines]
    ax1.legend(lines, labels, loc='upper right')

    plt.title("FedAvg on MedMNIST (pathmnist)")
    fig.tight_layout()
    plt.savefig("medmnist_fedavg_metrics.png")
    plt.show()
```

```python
import flwr as fl
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split, Subset
import torchvision.transforms as transforms
import numpy as np
import matplotlib.pyplot as plt
import copy
from medmnist import INFO
import medmnist

# Model for MedMNIST
class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        self.network = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),
            nn.Flatten(),
            nn.Linear(64 * 7 * 7, 128),
            nn.ReLU(),
            nn.Linear(128, 9)
        )

    def forward(self, x):
        return self.network(x)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load Dataset
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[.5], std=[.5])
])

info = INFO["pathmnist"]
DataClass = getattr(medmnist, info["python_class"])
trainset = DataClass(split="train", download=True, transform=transform)
testset = DataClass(split="test", download=True, transform=transform)
testloader = DataLoader(testset, batch_size=32)

#  Partitioning
def get_client_dataloader(client_id, num_clients, batch_size=32):
    total_size = len(trainset)
    split_size = total_size // num_clients
    indices = list(range(total_size))
    np.random.seed(42)
    np.random.shuffle(indices)
    client_indices = indices[client_id * split_size : (client_id + 1) * split_size]
    subset = Subset(trainset, client_indices)
    train_len = int(0.8 * len(subset))
    val_len = len(subset) - train_len
    train_data, val_data = random_split(subset, [train_len, val_len])
    return (
        DataLoader(train_data, batch_size=batch_size, shuffle=True),
        DataLoader(val_data, batch_size=batch_size, shuffle=False),
    )

# Flower Client with SCAFFOLD
class ScaffoldClient(fl.client.NumPyClient):
    def __init__(self, cid, c_global, c_local_dict):
        self.cid = int(cid)
        self.model = SimpleCNN().to(DEVICE)
        self.trainloader, self.valloader = get_client_dataloader(self.cid, num_clients=5)
        self.c_global = c_global
        self.c_local = c_local_dict[self.cid]

    def get_parameters(self, config):
        return [val.cpu().numpy() for _, val in self.model.state_dict().items()]

    def set_parameters(self, parameters):
        state_dict = self.model.state_dict()
        for k, v in zip(state_dict.keys(), parameters):
            state_dict[k] = torch.tensor(v)
        self.model.load_state_dict(state_dict)

    def fit(self, parameters, config):
        self.set_parameters(parameters)
        self.model.train()
        optimizer = optim.SGD(self.model.parameters(), lr=0.01)
        criterion = nn.CrossEntropyLoss()
        local_epochs = 1
        old_weights = copy.deepcopy(self.model.state_dict())
        for _ in range(local_epochs):
            for x, y in self.trainloader:
                x, y = x.to(DEVICE), y.squeeze().long().to(DEVICE)
                optimizer.zero_grad()
                output = self.model(x)
                loss = criterion(output, y)
                loss.backward()
                for name, param in self.model.named_parameters():
                    if param.grad is not None:
                        param.grad += self.c_global[name].to(DEVICE) - self.c_local[name].to(DEVICE)
                optimizer.step()
        new_weights = self.model.state_dict()
        delta_weights = {k: new_weights[k] - old_weights[k] for k in new_weights}
        for k in self.c_local:
            self.c_local[k] = self.c_local[k] - self.c_global[k] + delta_weights[k] / (local_epochs * 0.01)
        return self.get_parameters({}), len(self.trainloader.dataset), {}

    def evaluate(self, parameters, config):
        self.set_parameters(parameters)
        self.model.eval()
        correct, total, loss = 0, 0, 0
        criterion = nn.CrossEntropyLoss()
        with torch.no_grad():
            for x, y in self.valloader:
                x, y = x.to(DEVICE), y.squeeze().long().to(DEVICE)
                outputs = self.model(x)
                loss += criterion(outputs, y).item()
                _, predicted = torch.max(outputs.data, 1)
                total += y.size(0)
                correct += (predicted == y).sum().item()
        return loss / len(self.valloader), len(self.valloader.dataset), {"accuracy": correct / total}

# Control Variates
c_global = {k: torch.zeros_like(v) for k, v in SimpleCNN().state_dict().items()}
c_local_dict = {i: {k: torch.zeros_like(v) for k, v in SimpleCNN().state_dict().items()} for i in range(5)}

# Client Factory
def client_fn(cid: str):
    return ScaffoldClient(cid, c_global, c_local_dict).to_client()

# Evaluation
test_losses, test_accuracies = [], []

def evaluate_fn(server_round, parameters, config):
    model = SimpleCNN().to(DEVICE)
    state_dict = model.state_dict()
    for k, v in zip(state_dict.keys(), parameters):
        state_dict[k] = torch.tensor(v)
    model.load_state_dict(state_dict)
    model.eval()
    criterion = nn.CrossEntropyLoss()
    correct, total, loss = 0, 0, 0
    with torch.no_grad():
        for x, y in testloader:
            x, y = x.to(DEVICE), y.squeeze().long().to(DEVICE)
            output = model(x)
            loss += criterion(output, y).item()
            _, predicted = torch.max(output.data, 1)
            total += y.size(0)
            correct += (predicted == y).sum().item()
    avg_loss = loss / len(testloader)
    accuracy = correct / total
    test_losses.append(avg_loss)
    test_accuracies.append(accuracy)
    print(f"[Round {server_round}] Test Loss: {avg_loss:.4f}, Accuracy: {accuracy:.4f}")
    return avg_loss, {"test_accuracy": accuracy}

# Simulation
if __name__ == "__main__":
    initial_model = SimpleCNN()
    initial_parameters = fl.common.ndarrays_to_parameters(
        [val.cpu().numpy() for _, val in initial_model.state_dict().items()]
    )
    fl.simulation.start_simulation(
        client_fn=client_fn,
        num_clients=5,
        config=fl.server.ServerConfig(num_rounds=100),
        strategy=fl.server.strategy.FedAvg(
            fraction_fit=1.0,
            fraction_evaluate=1.0,
            min_fit_clients=5,
            min_evaluate_clients=5,
            min_available_clients=5,
            evaluate_fn=evaluate_fn,
            initial_parameters=initial_parameters,
        ),
    )
    # Save results
    np.save("scaffold_medmnist_test_accuracies.npy", np.array(test_accuracies))
    np.save("scaffold_medmnist_test_losses.npy", np.array(test_losses))

    # Plot
    rounds = list(range(1, len(test_losses) + 1))
    fig, ax1 = plt.subplots(figsize=(10, 5))
    ax1.set_xlabel("Round")
    ax1.set_ylabel("Test Loss", color="red")
    ax1.plot(rounds, test_losses, color="red", marker="o")
    ax1.tick_params(axis="y", labelcolor="red")

    ax2 = ax1.twinx()
    ax2.set_ylabel("Test Accuracy", color="blue")
    ax2.plot(rounds, test_accuracies, color="blue", marker="x")
    ax2.tick_params(axis="y", labelcolor="blue")

    plt.title("SCAFFOLD on MedMNIST")
    plt.tight_layout()
    plt.savefig("scaffold_medmnist_loss_accuracy.png")
    plt.show()
```
