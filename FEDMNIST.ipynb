```python
import flwr as fl
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader, random_split, Subset
import numpy as np
import matplotlib.pyplot as plt

# 1. Model Definition (MNIST: 1 channel, 28x28, 10 classes)
class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
        )
        self.fc = nn.Sequential(
            nn.Flatten(),
            nn.Linear(64 * 7 * 7, 128),
            nn.ReLU(),
            nn.Linear(128, 10)  # MNIST has 10 digit classes
        )

    def forward(self, x):
        x = self.conv(x)
        x = self.fc(x)
        return x

# 2. Device Setup
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 3. Load MNIST Dataset (used as FedMNIST)
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])

trainset = torchvision.datasets.MNIST(root="./data", train=True, download=True, transform=transform)
testset = torchvision.datasets.MNIST(root="./data", train=False, download=True, transform=transform)

def get_client_dataloader(client_id, num_clients, batch_size=32):
    total_size = len(trainset)
    split_size = total_size // num_clients
    indices = list(range(client_id * split_size, (client_id + 1) * split_size))
    subset = Subset(trainset, indices)
    train_len = int(0.8 * len(subset))
    val_len = len(subset) - train_len
    train_data, val_data = random_split(subset, [train_len, val_len])
    return (
        DataLoader(train_data, batch_size=batch_size, shuffle=True),
        DataLoader(val_data, batch_size=batch_size, shuffle=False),
    )

testloader = DataLoader(testset, batch_size=32)

# 4. Flower Client
class FlowerClient(fl.client.NumPyClient):
    def __init__(self, cid):
        self.cid = int(cid)
        self.model = SimpleCNN().to(DEVICE)
        self.trainloader, self.valloader = get_client_dataloader(self.cid, num_clients=5)

    def get_parameters(self, config):
        return [val.cpu().numpy() for val in self.model.state_dict().values()]

    def set_parameters(self, parameters):
        state_dict = self.model.state_dict()
        for k, v in zip(state_dict.keys(), parameters):
            state_dict[k] = torch.tensor(v)
        self.model.load_state_dict(state_dict)

    def fit(self, parameters, config):
        self.set_parameters(parameters)
        self.model.train()
        optimizer = optim.Adam(self.model.parameters(), lr=0.001)
        criterion = nn.CrossEntropyLoss()
        total_loss = 0

        for epoch in range(1):
            for x, y in self.trainloader:
                x, y = x.to(DEVICE), y.to(DEVICE)
                optimizer.zero_grad()
                out = self.model(x)
                loss = criterion(out, y)
                loss.backward()
                optimizer.step()
                total_loss += loss.item()

        avg_loss = total_loss / len(self.trainloader)
        return self.get_parameters({}), len(self.trainloader.dataset), {"train_loss": avg_loss}

    def evaluate(self, parameters, config):
        self.set_parameters(parameters)
        self.model.eval()
        criterion = nn.CrossEntropyLoss()
        loss = 0
        correct = 0
        total = 0

        with torch.no_grad():
            for x, y in self.valloader:
                x, y = x.to(DEVICE), y.to(DEVICE)
                out = self.model(x)
                loss += criterion(out, y).item()
                preds = torch.argmax(out, dim=1)
                correct += (preds == y).sum().item()
                total += y.size(0)

        val_loss = loss / len(self.valloader)
        accuracy = correct / total
        return val_loss, len(self.valloader.dataset), {"val_accuracy": accuracy}

# 5. Client Factory
def client_fn(cid: str):
    return FlowerClient(cid).to_client()

# 6. Server Evaluation
test_losses, test_accuracies = [], []

def evaluate_fn(server_round, parameters, config):
    model = SimpleCNN().to(DEVICE)
    state_dict = model.state_dict()
    for k, v in zip(state_dict.keys(), parameters):
        state_dict[k] = torch.tensor(v)
    model.load_state_dict(state_dict)
    model.eval()

    criterion = nn.CrossEntropyLoss()
    loss = 0
    correct = 0
    total = 0

    with torch.no_grad():
        for x, y in testloader:
            x, y = x.to(DEVICE), y.to(DEVICE)
            out = model(x)
            loss += criterion(out, y).item()
            preds = torch.argmax(out, dim=1)
            correct += (preds == y).sum().item()
            total += y.size(0)

    avg_loss = loss / len(testloader)
    accuracy = correct / total

    test_losses.append(avg_loss)
    test_accuracies.append(accuracy)

    print(f"Round {server_round}: Test loss {avg_loss:.4f}, accuracy {accuracy:.4f}")
    return avg_loss, {"test_accuracy": accuracy}

# 7. Start Simulation
if __name__ == "__main__":
    model = SimpleCNN()
    initial_weights = [val.cpu().numpy() for val in model.state_dict().values()]
    initial_parameters = fl.common.ndarrays_to_parameters(initial_weights)

    fl.simulation.start_simulation(
        client_fn=client_fn,
        num_clients=5,
        config=fl.server.ServerConfig(num_rounds=100),
        strategy=fl.server.strategy.FedAdam(
            eta=0.001,
            eta_l=0.001,
            beta_1=0.9,
            beta_2=0.99,
            tau=1e-9,
            fraction_fit=1.0,
            fraction_evaluate=1.0,
            min_fit_clients=5,
            min_evaluate_clients=5,
            min_available_clients=5,
            evaluate_fn=evaluate_fn,
            initial_parameters=initial_parameters,
        ),
    )

    # 8. Save and Plot Results
    rounds = list(range(1, len(test_losses) + 1))
    plt.figure(figsize=(10, 5))
    plt.plot(rounds, test_losses, label="Test Loss", marker="o")
    plt.plot(rounds, test_accuracies, label="Test Accuracy", marker="o")
    plt.xlabel("Round")
    plt.ylabel("Metric")
    plt.title("FedAdam on FedMNIST")
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.savefig("fedadam_fedmnist_plot.png")
    plt.show()

    # Save arrays
    np.save("fedadam_fedmnist_losses.npy", np.array(test_losses))
    np.save("fedadam_fedmnist_accuracies.npy", np.array(test_accuracies))
```

```python
import flwr as fl
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision.datasets import MNIST
import torchvision.transforms as transforms
from torch.utils.data import DataLoader, Subset
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
import random

# Model
class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        self.network = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),
            nn.Flatten(),
            nn.Linear(64 * 7 * 7, 128),
            nn.ReLU(),
            nn.Linear(128, 10)  # MNIST: 10 digits
        )

    def forward(self, x):
        return self.network(x)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Dataset loading and transformation
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])

trainset = MNIST(root="./data", train=True, download=True, transform=transform)
testset = MNIST(root="./data", train=False, download=True, transform=transform)

# Simulate non-IID split: each client gets data from only 2 digits
def create_non_iid_partitions(dataset, num_clients=5, num_classes_per_client=2):
    class_indices = defaultdict(list)
    for idx, (_, label) in enumerate(dataset):
        class_indices[label].append(idx)

    client_indices = [[] for _ in range(num_clients)]
    available_classes = list(range(10))
    random.seed(42)

    for client_id in range(num_clients):
        client_classes = random.sample(available_classes, num_classes_per_client)
        for cls in client_classes:
            num_samples = len(class_indices[cls]) // num_clients
            chosen = class_indices[cls][:num_samples]
            client_indices[client_id].extend(chosen)
            class_indices[cls] = class_indices[cls][num_samples:]
    return client_indices

client_partitions = create_non_iid_partitions(trainset, num_clients=5)

def get_client_dataloader(client_id, batch_size=32):
    indices = client_partitions[client_id]
    subset = Subset(trainset, indices)
    split = int(0.8 * len(subset))
    train_split, val_split = torch.utils.data.random_split(subset, [split, len(subset) - split])
    return (
        DataLoader(train_split, batch_size=batch_size, shuffle=True),
        DataLoader(val_split, batch_size=batch_size, shuffle=False)
    )

testloader = DataLoader(testset, batch_size=32)

# Flower Client
class FlowerClient(fl.client.NumPyClient):
    def __init__(self, cid):
        self.cid = int(cid)
        self.model = SimpleCNN().to(DEVICE)
        self.trainloader, self.valloader = get_client_dataloader(self.cid)

    def get_parameters(self, config):
        return [val.cpu().numpy() for val in self.model.state_dict().values()]

    def set_parameters(self, parameters):
        state_dict = self.model.state_dict()
        for k, v in zip(state_dict.keys(), parameters):
            state_dict[k] = torch.tensor(v)
        self.model.load_state_dict(state_dict)

    def fit(self, parameters, config):
        self.set_parameters(parameters)
        self.model.train()
        optimizer = optim.SGD(self.model.parameters(), lr=0.01)
        criterion = nn.CrossEntropyLoss()
        for x, y in self.trainloader:
            x, y = x.to(DEVICE), y.to(DEVICE)
            optimizer.zero_grad()
            out = self.model(x)
            loss = criterion(out, y)
            loss.backward()
            optimizer.step()
        return self.get_parameters({}), len(self.trainloader.dataset), {}

    def evaluate(self, parameters, config):
        self.set_parameters(parameters)
        self.model.eval()
        criterion = nn.CrossEntropyLoss()
        loss, correct, total = 0, 0, 0
        with torch.no_grad():
            for x, y in self.valloader:
                x, y = x.to(DEVICE), y.to(DEVICE)
                out = self.model(x)
                loss += criterion(out, y).item()
                preds = torch.argmax(out, dim=1)
                correct += (preds == y).sum().item()
                total += y.size(0)
        return loss / len(self.valloader), len(self.valloader.dataset), {"val_accuracy": correct / total}

def client_fn(cid: str):
    return FlowerClient(cid).to_client()

# Evaluation
test_losses, test_accuracies = [], []

def evaluate_fn(server_round, parameters, config):
    model = SimpleCNN().to(DEVICE)
    state_dict = model.state_dict()
    for k, v in zip(state_dict.keys(), parameters):
        state_dict[k] = torch.tensor(v)
    model.load_state_dict(state_dict)
    model.eval()

    criterion = nn.CrossEntropyLoss()
    correct, total, loss = 0, 0, 0

    with torch.no_grad():
        for x, y in testloader:
            x, y = x.to(DEVICE), y.to(DEVICE)
            out = model(x)
            loss += criterion(out, y).item()
            pred = torch.argmax(out, dim=1)
            correct += (pred == y).sum().item()
            total += y.size(0)

    accuracy = correct / total
    avg_loss = loss / len(testloader)
    test_losses.append(avg_loss)
    test_accuracies.append(accuracy)

    print(f"Round {server_round}: Test Loss = {avg_loss:.4f}, Accuracy = {accuracy:.4f}")
    return avg_loss, {"test_accuracy": accuracy}

# Run FL simulation
if __name__ == "__main__":
    model = SimpleCNN()
    initial_weights = [val.cpu().numpy() for val in model.state_dict().values()]
    initial_parameters = fl.common.ndarrays_to_parameters(initial_weights)

    fl.simulation.start_simulation(
        client_fn=client_fn,
        num_clients=5,
        config=fl.server.ServerConfig(num_rounds=100),
        strategy=fl.server.strategy.FedAvg(
            fraction_fit=1.0,
            fraction_evaluate=1.0,
            min_fit_clients=5,
            min_evaluate_clients=5,
            min_available_clients=5,
            evaluate_fn=evaluate_fn,
            initial_parameters=initial_parameters,
        ),
    )

    # Save metrics
    np.save("fedavg_fedmnist_test_accuracies.npy", np.array(test_accuracies))
    np.save("fedavg_fedmnist_test_losses.npy", np.array(test_losses))

    # Plot
    rounds = list(range(1, len(test_losses) + 1))
    fig, ax1 = plt.subplots(figsize=(10, 5))

    ax1.set_xlabel("Round")
    ax1.set_ylabel("Test Loss")
    line1, = ax1.plot(rounds, test_losses, marker='o', label="Test Loss")
    ax1.tick_params(axis='y')

    ax2 = ax1.twinx()
    ax2.set_ylabel("Test Accuracy")
    line2, = ax2.plot(rounds, test_accuracies, marker='x', label="Test Accuracy")
    ax2.tick_params(axis='y')

    ax1.legend([line1, line2], ["Test Loss", "Test Accuracy"], loc='upper right')
    plt.title("FedAvg on Simulated FedMNIST")
    fig.tight_layout()
    plt.savefig("fedavg_fedmnist_simulated_metrics.png")
    plt.show()
```

SCAFFOLD WITH FEDMNIST- 1 LOCAL EPOCH WITH 100 ROUNDS

```python
import flwr as fl
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as transforms
from torchvision.datasets import MNIST
from torch.utils.data import DataLoader, Subset, random_split
import numpy as np
import matplotlib.pyplot as plt
import copy
import random
from collections import defaultdict

# Model
class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        self.network = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),
            nn.Flatten(),
            nn.Linear(64 * 7 * 7, 128),
            nn.ReLU(),
            nn.Linear(128, 10)
        )

    def forward(self, x):
        return self.network(x)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Transform
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])

# Load MNIST
trainset = MNIST(root="./data", train=True, download=True, transform=transform)
testset = MNIST(root="./data", train=False, download=True, transform=transform)
testloader = DataLoader(testset, batch_size=32)

# Non-IID partitioning
def create_non_iid_partitions(dataset, num_clients=5, num_classes_per_client=2):
    class_indices = defaultdict(list)
    for idx, (_, label) in enumerate(dataset):
        class_indices[label].append(idx)

    client_indices = [[] for _ in range(num_clients)]
    available_classes = list(range(10))
    random.seed(42)

    for client_id in range(num_clients):
        client_classes = random.sample(available_classes, num_classes_per_client)
        for cls in client_classes:
            num_samples = len(class_indices[cls]) // num_clients
            chosen = class_indices[cls][:num_samples]
            client_indices[client_id].extend(chosen)
            class_indices[cls] = class_indices[cls][num_samples:]
    return client_indices

client_partitions = create_non_iid_partitions(trainset, num_clients=5)

def get_client_dataloader(client_id, batch_size=32):
    indices = client_partitions[client_id]
    subset = Subset(trainset, indices)
    split = int(0.8 * len(subset))
    train_data, val_data = random_split(subset, [split, len(subset) - split])
    return (
        DataLoader(train_data, batch_size=batch_size, shuffle=True),
        DataLoader(val_data, batch_size=batch_size, shuffle=False)
    )

# SCAFFOLD Client
class ScaffoldClient(fl.client.NumPyClient):
    def __init__(self, cid, c_global, c_local_dict):
        self.cid = int(cid)
        self.model = SimpleCNN().to(DEVICE)
        self.trainloader, self.valloader = get_client_dataloader(self.cid)
        self.c_global = c_global
        self.c_local = c_local_dict[self.cid]

    def get_parameters(self, config):
        return [val.cpu().numpy() for _, val in self.model.state_dict().items()]

    def set_parameters(self, parameters):
        state_dict = self.model.state_dict()
        for k, v in zip(state_dict.keys(), parameters):
            state_dict[k] = torch.tensor(v)
        self.model.load_state_dict(state_dict)

    def fit(self, parameters, config):
        self.set_parameters(parameters)
        self.model.train()

        optimizer = optim.SGD(self.model.parameters(), lr=0.01)
        criterion = nn.CrossEntropyLoss()
        local_epochs = 1
        old_weights = copy.deepcopy(self.model.state_dict())

        for _ in range(local_epochs):
            for x, y in self.trainloader:
                x, y = x.to(DEVICE), y.to(DEVICE)
                optimizer.zero_grad()
                output = self.model(x)
                loss = criterion(output, y)
                loss.backward()
                for name, param in self.model.named_parameters():
                    if param.grad is not None:
                        param.grad += self.c_global[name].to(DEVICE) - self.c_local[name].to(DEVICE)
                optimizer.step()

        new_weights = self.model.state_dict()
        delta_weights = {k: new_weights[k] - old_weights[k] for k in new_weights}
        for k in self.c_local:
            self.c_local[k] = self.c_local[k] - self.c_global[k] + delta_weights[k] / (local_epochs * 0.01)

        return self.get_parameters({}), len(self.trainloader.dataset), {}

    def evaluate(self, parameters, config):
        self.set_parameters(parameters)
        self.model.eval()
        correct, total, loss = 0, 0, 0
        criterion = nn.CrossEntropyLoss()
        with torch.no_grad():
            for x, y in self.valloader:
                x, y = x.to(DEVICE), y.to(DEVICE)
                output = self.model(x)
                loss += criterion(output, y).item()
                _, predicted = torch.max(output.data, 1)
                correct += (predicted == y).sum().item()
                total += y.size(0)
        return loss / len(self.valloader), len(self.valloader.dataset), {"accuracy": correct / total}

# Control Variates
c_global = {k: torch.zeros_like(v) for k, v in SimpleCNN().state_dict().items()}
c_local_dict = {i: {k: torch.zeros_like(v) for k, v in SimpleCNN().state_dict().items()} for i in range(5)}

# Client Factory
def client_fn(cid: str):
    return ScaffoldClient(cid, c_global, c_local_dict).to_client()

# Evaluation Function
test_losses, test_accuracies = [], []

def evaluate_fn(server_round, parameters, config):
    model = SimpleCNN().to(DEVICE)
    state_dict = model.state_dict()
    for k, v in zip(state_dict.keys(), parameters):
        state_dict[k] = torch.tensor(v)
    model.load_state_dict(state_dict)
    model.eval()

    criterion = nn.CrossEntropyLoss()
    correct, total, loss = 0, 0, 0
    with torch.no_grad():
        for x, y in testloader:
            x, y = x.to(DEVICE), y.to(DEVICE)
            output = model(x)
            loss += criterion(output, y).item()
            _, predicted = torch.max(output.data, 1)
            correct += (predicted == y).sum().item()
            total += y.size(0)

    avg_loss = loss / len(testloader)
    accuracy = correct / total
    test_losses.append(avg_loss)
    test_accuracies.append(accuracy)
    print(f"[Round {server_round}] Test Loss: {avg_loss:.4f}, Accuracy: {accuracy:.4f}")
    return avg_loss, {"test_accuracy": accuracy}

# Simulation
if __name__ == "__main__":
    initial_model = SimpleCNN()
    initial_parameters = fl.common.ndarrays_to_parameters(
        [val.cpu().numpy() for _, val in initial_model.state_dict().items()]
    )

    fl.simulation.start_simulation(
        client_fn=client_fn,
        num_clients=5,
        config=fl.server.ServerConfig(num_rounds=100),
        strategy=fl.server.strategy.FedAvg(
            fraction_fit=1.0,
            fraction_evaluate=1.0,
            min_fit_clients=5,
            min_evaluate_clients=5,
            min_available_clients=5,
            evaluate_fn=evaluate_fn,
            initial_parameters=initial_parameters,
        ),
    )

    # Save metrics
    np.save("scaffold_fedmnist_test_accuracies.npy", np.array(test_accuracies))
    np.save("scaffold_fedmnist_test_losses.npy", np.array(test_losses))

    # Plot
    rounds = list(range(1, len(test_losses) + 1))
    fig, ax1 = plt.subplots(figsize=(10, 5))
    ax1.set_xlabel("Round")
    ax1.set_ylabel("Test Loss", color="red")
    ax1.plot(rounds, test_losses, color="red", marker="o")
    ax1.tick_params(axis="y", labelcolor="red")

    ax2 = ax1.twinx()
    ax2.set_ylabel("Test Accuracy", color="blue")
    ax2.plot(rounds, test_accuracies, color="blue", marker="o")
    ax2.tick_params(axis="y", labelcolor="blue")

    plt.title("SCAFFOLD on Simulated FedMNIST")
    fig.tight_layout()
    plt.savefig("scaffold_fedmnist_loss_accuracy.png")
    plt.show()
```
